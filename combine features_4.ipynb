```python
#_____________________________________________________ for screen

def p_num_from_tr_id(tracking_record, p_tr_id):
    for num in range(1,11):
        if tracking_record[f'p{num}_player_id'] == p_tr_id:
            return num

def transform_coord_pair(x,y):
    if x > 47 :
        y = 50 - y
        x = 94 - x
    y = 250 * (1 - (y - 0)/(50 - 0)) + -250 * ((y - 0)/(50 - 0))
    x = -47.5 * (1 - (x - 0)/(47 - 0)) + 422.5 * ((x - 0)/(47 - 0))

    return x,y

def calc_dist(x_1, y_1, x_2, y_2):
    return np.sqrt((x_2 - x_1) ** 2 + (y_2 - y_1) ** 2)


def get_dists_to_ball(df):
    
    #avg_dists=[]
    min_def_dists=[]
    min_off_dists=[]
    second_min_def_dists = []
    second_min_off_dists = []
    ballhandlers = []
    on_ball_defs = []
    screeners = []
    screener_defs = []

    for idx,val in df.iterrows():
        b_x,b_y = val['b_x_x'],val['b_y_x']
        shooter_team = val['SHOOTER_TEAM_ID']
        off_players = []
        def_players = []
        off_dists = []
        def_dists = []
        for num in range(1,11):
            if val[f'p{num}_team_id_x'] != shooter_team:
                def_dist = calc_dist(b_x,b_y,val[f'p{num}_x_x'],val[f'p{num}_y_x'])
                def_players.append(f'P{num}_Name')
                def_dists.append(def_dist)
            if val[f'p{num}_team_id_x'] == shooter_team:
                off_dist = calc_dist(b_x,b_y,val[f'p{num}_x_x'],val[f'p{num}_y_x'])
                off_players.append(f'P{num}_Name') #P10_Name
                off_dists.append(off_dist)
        #avg_dist = sum(dists)/len(dists)
        off_tuple = list(zip(off_players,off_dists))
        off_tuple = np.array(off_tuple)
        def_tuple = list(zip(def_players,def_dists))
        def_tuple = np.array(def_tuple)
        sort_off = np.argsort(off_tuple[:,1])
        sort_def = np.argsort(def_tuple[:,1])
        min_off = off_tuple[sort_off[0],:]
        min_def = def_tuple[sort_def[0],:]
        second_off = off_tuple[sort_off[1],:]
        second_def = def_tuple[sort_def[1],:]
        
        #min_off_dists.append(min_off[1])
        min_def_dists.append(min_def[1])
        #second_min_off_dists.append(second_off[1])
        second_min_def_dists.append(second_def[1])
        ballhandlers.append(min_off[0])
        on_ball_defs.append(min_def[0])
        screeners.append(second_off[0])
        screener_defs.append(second_def[0])


        #avg_dists.append(avg_dist)
        #min_dists.append(min_dist)
    df['Ball_handler'] = ballhandlers
    df['On_ball_defender'] = on_ball_defs
    df['On_ball_defender_dist_to_ball'] = min_def_dists
    df['Screener'] = screeners
    df['Screener_defender'] = screener_defs
    df['Screener_defender_dist_to_ball'] = second_min_def_dists
    return df

#_____________________________________________________
```
