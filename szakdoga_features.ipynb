```python
filter_shots = shots.drop(columns=['GRID_TYPE','TEAM_NAME','EVENT_TYPE','SHOT_ATTEMPTED_FLAG','GAME_DATE','HTM','VTM'])

filter_shots.head(5)
```

```python
filter_shots.loc[(filter_shots['SHOT_DISTANCE'] < 3) & (filter_shots['SHOT_DISTANCE'] > -1),'SHOT_ZONE_RANGE']= '0-3'
filter_shots.loc[(filter_shots['SHOT_DISTANCE'] < 10) & (filter_shots['SHOT_DISTANCE'] >= 3),'SHOT_ZONE_RANGE']= '3-10'
filter_shots.loc[(filter_shots['SHOT_DISTANCE'] < 16) & (filter_shots['SHOT_DISTANCE'] >= 10),'SHOT_ZONE_RANGE']= '10-16'
filter_shots.loc[(filter_shots['SHOT_DISTANCE'] < 22) & (filter_shots['SHOT_DISTANCE'] >= 16),'SHOT_ZONE_RANGE']= '16-3P'
filter_shots.loc[(filter_shots['SHOT_DISTANCE'] >= 22),'SHOT_ZONE_RANGE']= '3P'
```

```python
filter_shots.head(5)
```

```python
clean_bballref = pd.read_csv('clean_bballref.csv')
```

```python
shot_and_stat = filter_shots.merge(clean_bballref, how='left', left_on = 'PLAYER_NAME', right_on= 'Name')
```

```python
shot_and_stat.head(5)
```

```python
shot_and_stat['MINUTES_REMAINING'] = shot_and_stat['MINUTES_REMAINING']*60
shot_and_stat['SECLEFT'] = shot_and_stat['MINUTES_REMAINING'] + shot_and_stat['SECONDS_REMAINING']
```

```python
shot_and_stat.head(5)
```

```python
shot_and_stat.to_csv('shot_and_stat.csv')
```
