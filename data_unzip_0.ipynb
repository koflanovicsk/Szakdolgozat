```python
import pandas
import os
import sys
import json

#import movement.config as CONFIG

data_path = 'D:\\Szakdolgozat_data\\nba-movement-data-master\\data\\JSON'#CONFIG.data.dir
csv_path = 'D:\\Szakdolgozat_data\\nba-movement-data-master\\data\\csv_fix'#CONFIG.data.movement.dir
files = os.listdir(data_path)

if not os.path.exists(csv_path):
    os.makedirs(csv_path)

count = 0
movement_headers = ["game_id","event_id","game_clock", "shot_clock", "quarter",\
                    "b_team_id", "b_player_id", "b_x", "b_y", "b_z",\
                        "p1_team_id", "p1_player_id", "p1_x", "p1_y", "p1_z",\
                        "p2_team_id", "p2_player_id", "p2_x", "p2_y", "p2_z",\
                        "p3_team_id", "p3_player_id", "p3_x", "p3_y", "p3_z",\
                        "p4_team_id", "p4_player_id", "p4_x", "p4_y", "p4_z",\
                        "p5_team_id", "p5_player_id", "p5_x", "p5_y", "p5_z",\
                        "p6_team_id", "p6_player_id", "p6_x", "p6_y", "p6_z",\
                        "p7_team_id", "p7_player_id", "p7_x", "p7_y", "p7_z",\
                        "p8_team_id", "p8_player_id", "p8_x", "p8_y", "p8_z",\
                        "p9_team_id", "p9_player_id", "p9_x", "p9_y", "p9_z",\
                        "p10_team_id", "p10_player_id", "p10_x", "p10_y", "p10_z", ]
for file in files:
    if '.json' not in file:
        continue
    try:
        count = count + 1
        file_data = open('%s/%s' % (data_path, file))
        game_id = file.replace('.json', '')
        data = json.load(file_data)
        events = data['events']
        moments = []

        for event in events:
            event_id = event['eventId']
            movement_data = event['moments']
            for moment in movement_data:
                mmnt = [game_id, event_id, moment[2], moment[3], moment[0]]
                for player in moment[5]:
                    for stat in player:
                        mmnt.append(stat)
                    #mmnt.append(stat for stat in player)
                moments.append(mmnt)

        # movement frame is complete for game
        movement = pandas.DataFrame(moments, columns=movement_headers)
        movement.to_csv('%s/%s.csv' % (csv_path, game_id), index=False)
        # movement.to_json('./data/json/' + game_id + '.json', orient='records')

        print('\n')
        print('\n')
        print(f'Finished collecting dataframe for Game ID: {game_id}')
        print(f'Completed : {count} games.')
    except Exception as e:
        print(f'Error in loading: {file} file, Error: {e}')

print('\n')
print('\n')
print('Finished collecting dataframes for all games.')
print(f'{count} games counted')
```
