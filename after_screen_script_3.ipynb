```python
meccsek = {}
for file in range(21500001, 21500152): #21500152
    try:
        track = pd.read_csv(r'D:\Szakdolgozat_data\nba-movement-data-master\data\csv_fix\00{0}.csv'.format(file))
        gamea = dist_notzero_nodup.loc[dist_notzero_nodup['Game_ID'] == file]
        game = gamea[['Game_ID','Event_Number','Frame_Number']]
        
        group = track.groupby('event_id',as_index = False)
        grouped_index = group.apply(lambda x: x.reset_index(drop = True)).reset_index()
        
        track_fin = grouped_index.drop('level_0',axis = 1)
        track_fin.rename(columns = {'level_1':'frame_number'}, inplace = True)
        meccsek[file] = pd.merge(game, track_fin,  how='inner', left_on=['Game_ID','Event_Number','Frame_Number'], right_on =['game_id','event_id','frame_number'])
    except OSError as e:
        continue
```

```python
screenstill150 = pd.concat(meccsek)
screenstill150
```
